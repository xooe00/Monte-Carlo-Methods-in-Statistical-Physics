```python
# 라이브러리 임포트
import numpy as np
import math as m
import random
import matplotlib.pyplot as plt
```

```python
z = 3

def g(x):
    result = np.where(x > z, 1, 0)
    return result
```

```python
def p(x):
    return 1 / np.sqrt(2*np.pi) * np.exp(-x**2/2)
```

```python
def q(x):
    return 1 / np.sqrt(2*np.pi) * np.exp(-(x-z)**2/2)
```

```python
def w(x):
    return p(x)/q(x)
```

```python
def func1(N):
    x_stars = np.random.normal(0, 1, N) # x* ~ N(0,1)

    result = g(x_stars)

    return sum(result)/N #, np.var(result)
```

```python
func1(10000)
```

```python
def func2(N):
    x_stars = np.random.normal(3, 1, N) # x* ~ N(3,1)

    result = g(x_stars)*w(x_stars)

    return sum(result)/N #, np.var(result)
```

```python
func2(10000)
```
