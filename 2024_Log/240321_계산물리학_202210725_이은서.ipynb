```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
P = np.array([[0.7, 0.2, 0.1], [0.3, 0.65, 0.05], [0.15, 0.35, 0.5]]) # 확률 행렬 P
```

```python
def func(N):
    w = 1 # 날씨 초기값: w=1
    X = [1] # 날씨 담는 리스트 S=1, C=2, R=3

    for _ in range(1, N):
        r = random.uniform(0, 1) # 랜덤 실수

        if w == 1: # 오늘 날씨가 S일 때
            if r < 0.7: # 0~0.7의 실수가 나오면 다음날은 S
                w = 1
            elif 0.7 <= r < 0.9: # 0.7~0.9의 실수가 나오면 다음날은 C
                w = 2
            else: # 0.9~1의 실수가 나오면 다음날은 R
                w = 3

        elif w == 2: # 오늘 날씨가 C일 때
            if r < 0.3: # 0~0.3의 실수가 나오면 다음날은 S
                w = 1
            elif 0.3 <= r < 0.95: # 0.3~0.95의 실수가 나오면 다음날은 C
                w = 2
            else:  # 0.95~1의 실수가 나오면 다음날은 R
                w = 3

        else: # 오늘 날씨가 R일 때
            if r < 0.15: # 0~0.15의 실수가 나오면 다음날은 S
                w = 1
            elif 0.15 <= r < 0.5: # 0.15~0.5의 실수가 나오면 다음날은 C
                w = 2
            else: # 0.5~1의 실수가 나오면 다음날은 R
                w = 3

        X.append(w)

    return X
```

```python
func(10)
```

```python
plt.plot(func(100))
```
