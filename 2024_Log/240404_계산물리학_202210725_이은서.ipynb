```python
import math
import random
import matplotlib.pyplot as plt
```

```python
# 확률밀도함수 p(x) 정의
def p(x):
    return 0.3/(1.5*math.sqrt(2*math.pi))*math.e**(-0.5*((x-(-3))/1.5)**2) + 0.7/(1.5*math.sqrt(2*math.pi))*math.e**(-0.5*((x-5)/1.5)**2)
```

```python
p(5)
```

```python
p(-3)
```

```python
def func(N):
    x_stars = []
    C = 4

    while len(x_stars) < N: # x*를 N개 구할 때까지 반복
        r = random.uniform(-8, 12) # -8~12 사이의 r
        u = random.uniform(0, 1) # 0~1 사이의 u
        if u <= p(r) / (C * 0.05): # u가 p(r)/Cq(r)보다 작거나 같을 경우
            x_stars.append(r) # r = x*
        else: # 아닐 경우
            pass # 새로운 r과 u를 구함
    return x_stars
```

```python
func(10)
```

```python
plt.hist(func(1000), bins=40)
plt.show()
```
