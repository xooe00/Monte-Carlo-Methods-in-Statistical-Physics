## 분포의 평균, 표준오차를 구하는 함수

```python
def func(N):
    n = int(N/10000) # N은 n의 1만 배
    x_list = np.random.uniform(0, 1, N) # 0~1 사이의 실수가 N개 담긴 X
    x_array = x_list.reshape(-1, n) # x를 n개씩 쪼개어 담고
    y_list = np.sum(x_array, axis=1) / n # x의 n개 평균인 y가 담긴 Y
    mean_y = np.mean(y_list) # Y의 평균
    error_y = np.sqrt(np.var(y_list)) # Y의 오차
    return y_list, mean_y, error_y
```

```python
# N = 10^5 일 때 Y
test, _, _ = func(100000)
plt.hist(test,100)
plt.show()
```

```python
Y1, v1, e1 = func(100000)
Y2, v2, e2 = func(1000000)
Y3, v3, e3 = func(10000000)
Y4, v4, e4 = func(100000000)
print(f'N = 10^5, variance = {v1}, error = {e1}')
print(f'N = 10^6, variance = {v2}, error = {e2}')
print(f'N = 10^7, variance = {v3}, error = {e3}')
print(f'N = 10^8, variance = {v4}, error = {e4}')
```

N이 100배 늘 때 오차가 10배 줄어드는 것을 확인할 수 있다.
$$\hatσ\propto\frac{1}{\sqrt{N}}$$

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8), sharex=True, sharey=True)

axs[0, 0].hist(Y1, bins=100)
axs[0, 0].set_title('N = 10^5')

axs[0, 1].hist(Y2, bins=100, color='green')
axs[0, 1].set_title('N = 10^6')

axs[1, 0].hist(Y3, bins=100, color='red')
axs[1, 0].set_title('N = 10^7')

axs[1, 1].hist(Y4, bins=100, color='purple')
axs[1, 1].set_title('N = 10^8')

plt.tight_layout()
plt.show()
```
