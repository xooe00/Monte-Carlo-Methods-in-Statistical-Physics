```python
# 라이브러리 임포트
import numpy as np
import math as m
import random
import matplotlib.pyplot as plt
```

```python
p = 0.4

def P(N):
    r = np.random.rand(N, 50) # 0~1 사이의 r을 N행 50열 뽑음

    X = np.where(r < p, 1, -1) # r<0.4인 경우 오른쪽, 그외의 경우 왼쪽

    result = np.sum(X, axis=1) # 열(50번)마다의 최종 이동 위치

    P = np.array(result[result>15])/N # X>15인 경우의 확률

    return np.mean(P) # 확률의 평균
```

```python
P(100000)
```

```python
q = 0.65

def P2(N):
    r = np.random.rand(N, 50) # 0~1 사이의 r을 N행 50열 뽑음

    X = np.where(r < q, 1, -1) # r<0.65인 경우 오른쪽, 그외의 경우 왼쪽

    P = np.sum(np.prod(np.where(X == 1, p/q, (1-p)/(1-q)), axis=1)[np.sum(X, axis=1) > 15])/N # 오른쪽인 경우 p/q, 왼쪽인 경우 (1-p)/(1-q)
    # X>15인 경우만 가중치의 곱을 구하여 N으로 나눈다

    return np.mean(P) # 확률의 평균
```

```python
P2(100000)
```
