```python
# 라이브러리 임포트
import numpy as np
import math as m
import random
import matplotlib.pyplot as plt
```

```python
def g(x):
    return m.e**(-0.5*x**2)
```

```python
a = 0
b = 10

def p(x):
    mask = (a < x) & (x < b)
    result = np.where(mask, 1/(b-a), 0)
    return result
```

```python
def q(x):
    return m.e**(-x)/(1-m.e**(-10))
```

```python
def w(x):
    return p(x)/q(x)
```

```python
def func1(N):
    x_stars = np.random.uniform(0, 10, N) # x* ~ U(0,1)

    result = g(x_stars)

    return sum(result)/N #, np.var(result)
```

```python
func1(10000)
```

```python
def func2(N):
    u = np.random.uniform(0, 1, N) # 0~1 사이의 u
    x_stars = -np.log(1-(1-np.exp(-10))*u) # 역변환법

    result = g(x_stars)*w(x_stars)

    return sum(result)/N #, np.var(result)
```

```python
func2(10000)
```
