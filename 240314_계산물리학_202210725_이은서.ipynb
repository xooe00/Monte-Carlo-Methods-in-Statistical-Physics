```python
def newfunc(N):
    result = []
    for i in range(1,N+1):
        x = random.uniform(0,1)
        y = math.sqrt(x)
        result.append(y)
    return result
```

```python
from scipy.special import beta, gamma
```

```python
def f2(x):
    a, b = 2, 2
    y = (gamma(a+b)/(gamma(a)*gamma(b))) * (x**(a-1)) * ((1-x)**(b-1))
    return float(y)
```

```python
def f3(x):
    y = math.sqrt(x)
    return float(y)
```

```python
def g(x):
    return float(3*x**2)
```

```python
def G1(N):
    sum = 0
    for i in range(1,N+1):
        r = random.uniform(0,1)
        sum += g(r)/f1(r)
    result = float(sum/N)
    return result
```

```python
import numpy
```

```python
def G2(N):
    sum = 0
    for i in range(1,N+1):
        r = numpy.random.beta(2,2)
        sum += g(r)/f2(r)
    result = float(sum/N)
    return result
```

```python
def G3(N):
    sum = 0
    for i in range(1,N+1):
        r = random.uniform(0,1)
        sum += g(f3(r))/(2*f3(r))
    result = float(sum/N)
    return result
```

```python
def V1(n):
    gi_list = []
    sig2_list = []
    gsum = 0.0
    result = 0.0

    for i in range(1,n+1):
        gi = float(G1(10000))
        gi_list.append(gi)
        gsum += gi
    gbar = gsum/n

    for data in gi_list:
        result += (data-gbar)**2

    return float(result/(n-1))
```
