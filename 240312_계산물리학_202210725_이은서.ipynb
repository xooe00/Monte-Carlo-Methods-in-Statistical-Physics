```python
def f(x):
    return float(1/2)
```

```python
def g(x):
    return float(3*x**2)
```

```python
def G(N):
    sum = 0
    for i in range(1,N+1):
        n = random.uniform(0,2)
        sum += g(n)/f(n)
    result = sum/N
    print(result)
```

```python
G(10000)
```

```python
def Eg(N):
    sum = 0
    for i in range(1,N+1):
        n = random.uniform(0,2)
        sum += g(n)
    result = sum/N
    print(result)
```
